# Pandas Data Visualization Exercise

This is just a quick exercise for you to review the various plots we showed earlier. Use **df3** to replicate the following plots. 

```python
import pandas as pd
import matplotlib.pyplot as plt
df3 = pd.read_csv('df3-1')
%matplotlib inline
```

```python
df3.info()
```

```python
df3.head()
```

**1. Recreate this scatter plot of b vs a. Note the color and size of the points. Also note the figure size. See if you can figure out how to stretch it in a similar fashion. Remeber back to your matplotlib lecture... (15 points)**

```python
fig, axes = plt.subplots(figsize=(8,2))
axes.scatter(df3['a'],df3['b'], c='red', s = 25, edgecolor='black')
axes.set_xlim([-0.2,1.2])
axes.set_ylim([-0.2,1.2])
axes.set_xlabel('a')
axes.set_ylabel('b');
```

**2. Create a histogram of the 'a' column. (15 points)**

```python

fig, axes = plt.subplots(1,1)
axes.hist(df3['a'], color='blue',edgecolor='black')
axes.set_ylabel('Frequency')
axes.set_xlim([0,1.0])
axes.set_ylim([0,70]);
```

**3. These plots are okay, but they don't look very polished. Use style sheets to set the style to 'ggplot' and redo the histogram from above. Also figure out how to add more bins to it. (15 points)***

```python
plt.style.use('ggplot')
fig, axes = plt.subplots(1,1)
axes.hist(df3['a'],bins = 25, alpha=0.5, edgecolor='white')
axes.set_ylabel('Frequency')
axes.set_xlim([0,1.0])
axes.set_ylim([0,35]);
```

**4. Create a boxplot comparing the a and b columns.(15points)**

```python

s = df3[['a','b']].plot.box()
s.set_ylim([0,1.0]);
```

**5. Create a kde plot of the 'd' column (15 points)**

```python
t = df3['d'].plot.kde()
t.set_ylim([0,1.2])
t.set_xlim([-0.5,1.5]);
```

**6. Figure out how to increase the linewidth and make the linestyle dashed. (Note: You would usually not dash a kde plot line) (15 points)**

```python
t = df3['d'].plot.kde(lw=4, ls='--')
t.set_ylim([0,1.2])
t.set_xlim([-0.5,1.5]);
```

**7. Create an area plot of all the columns for just the rows up to 30. (10 points)**

```python
area1= df3.plot.area(alpha=0.5)
area1.set_ylim([0,3.0])
area1.set_xlim([0,30])
```
